```python
import yfinance as yf
import matplotlib.pyplot as plt

# Símbolo da ação que você deseja consultar (por exemplo, AAPL para a Apple)
symbol = 'AAPL'

# Obtém os dados históricos dos últimos 5 dias
stock_data = yf.download(symbol, period="5d")

# Exibe os dados no console
print(stock_data)

# Cria um gráfico das cotações nos últimos 5 dias
plt.figure(figsize=(10, 6))
plt.plot(stock_data['Close'], marker='o')
plt.title(f'Cotação da Ação {symbol} nos Últimos 5 Dias')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')
plt.grid(True)
plt.show()
```

Essa versão deverá analisar os preços da ação e criar uuma linha de tendências nnos preços futuro

Altrei o código adicionando uma pergunta sobre a quantidade de dias

```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Solicita ao usuário que insira o símbolo do Fundo de Investimento Imobiliário (FII) desejado
symbol = input("Insira o símbolo do FII que deseja pesquisar (por exemplo, HGLG11.SA): ")

# Solicita ao usuário que insira a quantidade de dias para avaliar a ação
days_to_evaluate = int(input("Por quantos dias você deseja avaliar a ação? "))

# Obtém os dados históricos com base na quantidade de dias especificada
fii_data = yf.download(symbol, period=f"{days_to_evaluate}d")

# Verifica se os dados foram obtidos com sucesso
if fii_data.empty:
    print(f"Não foi possível obter dados para o FII com o símbolo {symbol}. Verifique se o símbolo está correto e tente novamente.")
else:
    # Exibe os dados no console
    print(fii_data)

    # Cria um gráfico das cotações nos últimos dias em linhas ao longo do tempo
    plt.figure(figsize=(10, 6))
    plt.plot(fii_data.index, fii_data['Close'], label='Preço de Fechamento', marker='o', linestyle='-')

    # Calcula a curva de tendência (regressão linear)
    x = np.arange(len(fii_data))
    slope, intercept, r_value, p_value, std_err = stats.linregress(x, fii_data['Close'])
    trendline = intercept + slope * x
    plt.plot(fii_data.index, trendline, color='red', label='Curva de Tendência', linestyle='--')

    plt.title(f'Cotação do FII {symbol} nos Últimos {days_to_evaluate} Dias com Curva de Tendência')
    plt.xlabel('Data')
    plt.ylabel('Preço de Fechamento')
    plt.grid(True)
    plt.legend()
    plt.xticks(rotation=45)
    plt.show()
```
